```python
class VirtualMemoryMgmt:
  def __init__(self, filename):
    self.filename = filename
    f = open(filename, "r")
    a = f.readline()
    page_number, frame_number, window_size, string_num = a.split(" ")
    self.page_number = page_number
    self.frame_number = frame_number
    self.window_size = window_size
    self.string_num = string_num
    b = f.readline()
    string_list = b.split()
    self.string_list = string_list

  def MIN(self, frame_number, string_num, string_list):
    print("MIN")
    import sys
    page_fault = 0
    memory_state = []

    for i in range(int(string_num)):
      if string_list[0] in memory_state:  #이미 memory에 할당된 경우
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state)
      elif len(memory_state) < int(frame_number):  #memory에 빈 공간이 있는 경우
        memory_state.append(string_list[0])
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1
      else:  #replacement strategies
        choose1 = {}
        for j in memory_state:
          try:
            index = string_list.index(j)
            choose1[str(j)] = index
          except:
            choose1[str(j)] = sys.maxsize  #최대값으로 설정
        choose1 = dict(choose1.items())
        choose2 = list(choose1.values())
        pagenum = max(choose2)  #tie breaking: page with smallest page number
        outpagenum = list(choose1.keys())[choose2.index(pagenum)]
        memory_state[int(memory_state.index(str(outpagenum)))] = string_list[0]
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1

    print("MIN 총 page fault 발생 횟수:", page_fault)
    print("---------------------------------------------------")

  def FIFO(self, frame_number, string_num, string_list):
    print("FIFO")
    page_fault = 0
    memory_state = []
    fifo_list = []

    for i in range(int(string_num)):
      if string_list[0] in memory_state:  #이미 memory에 할당된 경우
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state)
      elif len(memory_state) < int(frame_number):  #memory에 빈 공간이 있는 경우
        memory_state.append(string_list[0])
        fifo_list.append(string_list[0])
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1
      else:  #replacement strategies
        memory_state[memory_state.index(str(fifo_list[0]))] = string_list[0]
        fifo_list.remove(fifo_list[0])
        fifo_list.append(string_list[0])
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1

    print("FIFO 총 page fault 발생 횟수:", page_fault)
    print("---------------------------------------------------")

  def LRU(self, frame_number, string_num, string_list):
    print("LRU")
    page_fault = 0
    memory_state = []
    referenced_list = []

    for i in range(int(string_num)):
      if string_list[0] in memory_state:  #이미 memory에 할당된 경우
        referenced_list.remove(string_list[0])
        referenced_list.append(string_list[0])
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state)
      elif len(memory_state) < int(frame_number):  #memory에 빈 공간이 있는 경우
        memory_state.append(string_list[0])
        referenced_list.append(string_list[0])
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1
      else:  #replacement strategies
        memory_state[memory_state.index(referenced_list[0])] = string_list[0]
        referenced_list.remove(referenced_list[0])
        referenced_list.append(string_list[0])
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1

    print("LRU 총 page fault 발생 횟수:", page_fault)
    print("---------------------------------------------------")

  def LFU(self, frame_number, string_num, string_list):
    print("LFU")
    page_fault = 0
    memory_state = []
    referenced_list = []
    referenced_dict = {}

    for i in range(int(string_num)):
      if string_list[0] in memory_state:  #이미 memory에 할당된 경우
        referenced_list.remove(string_list[0])
        referenced_list.append(string_list[0])
        referenced_dict[string_list[0]] = int(referenced_dict[string_list[0]]) + 1
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state)
      elif len(memory_state) < int(frame_number):  #memory에 빈 공간이 있는 경우
        memory_state.append(string_list[0])
        referenced_list.append(string_list[0])
        referenced_dict[string_list[0]] = 1
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1
      else:  #replacement strategies
        choose = {}
        for h in memory_state:
          choose[h] = referenced_dict[h]
        reference_count = list(choose.values())
        minpage_num = reference_count.count(min(reference_count))
        if minpage_num > 1:  #tie breaking
          memory_state[memory_state.index(referenced_list[0])] = string_list[0]
          referenced_list.remove(referenced_list[0])
        else:
          outpagenum = list(choose.keys())[reference_count.index(min(reference_count))]
          memory_state[int(memory_state.index(str(outpagenum)))] = string_list[0]
          referenced_list.remove(outpagenum)   #여기 고쳤다
        referenced_list.append(string_list[0])
        if string_list[0] in referenced_dict.keys():
          referenced_dict[string_list[0]] = int(referenced_dict[string_list[0]]) + 1
        else:
          referenced_dict[string_list[0]] = 1
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), memory_state, "page fault")
        page_fault += 1

    print("LFU 총 page fault 발생 횟수:", page_fault)
    print("---------------------------------------------------")

  def WS(self, page_number, window_size, string_num, string_list):
    print("Working Set")
    referenced_string = []
    memory_state = [[]]
    page_fault = 0
    for i in range(int(string_num)):
      if int(string_list[0]) in memory_state[-1]:  #이미 memory에 할당된 경우
        working_set = referenced_string[-1*int(window_size):]
        working_set.append(int(string_list[0]))
        memory_state.append(working_set)
        referenced_string.append(int(string_list[0]))
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), set(memory_state[-1]))
      else:  
        working_set = referenced_string[-1*int(window_size):]
        working_set.append(int(string_list[0]))
        memory_state.append(working_set)
        referenced_string.append(int(string_list[0]))
        string_list.remove(string_list[0])
        print("time %d memory state: "%(i+1), set(memory_state[-1]), "page fault")
        page_fault += 1
    print("Working Set 총 page fault 발생 횟수:", page_fault)
    print("---------------------------------------------------")
```
